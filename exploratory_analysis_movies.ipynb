```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Reading in our data
# Data from https://www.kaggle.com/danielgrijalvas/movies
from google.colab import files
uploaded = files.upload()
```

```python
import io

df = pd.read_csv(io.BytesIO(uploaded['movies.csv']))
```

```python
df
```

```python
# We will loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Exploring the basics of our data

print(df.dtypes)
```

```python
# Time to visualize
plt.figure(figsize = (15,8))
df.boxplot(['gross'])
```

This tells us that the gross of movies follows a fat-tail distribution.

```python
# Duplicate data serves no purpose in this set so we will drop them

df.drop_duplicates()
```

```python
# Taking a look to see if our data matches expectations

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
plt.figure(figsize = (15,8))
sns.regplot(x = "gross", y = "budget", data = df)
```

Spending more money on a movie nets more money. Good News!

```python
plt.figure(figsize = (15,8))
sns.regplot(x = "score", y = "gross", data = df)
```

The gross income of a movie is not linearly correlated with its IMDB score.

```python
# Correlation Matrix between all numeric columns

df.corr(method = 'pearson')
```

```python
df.corr(method = 'kendall')
```

```python
df.corr(method = 'spearman')
```

```python
correlation_matrix = df.corr()

plt.figure(figsize = (15,8))

sns.heatmap(correlation_matrix, annot = True)

plt.title("Numeric Columns")

plt.show()
```

```python
# We can look at non-numeric column correlations by using factorize

df.apply(lambda x: x.factorize()[0]).corr(method = 'pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method = 'pearson')

plt.figure(figsize = (15,8))

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix (all data)")

plt.show()
```

```python
# Lets print the pairs that have strong correlation

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values(kind = "quicksort")

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

```python
# Calculating gross sum ourselves for the top 15 companies

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
# Lets make a year columns so we can look at breakout years for companies

df['Year'] = df['released'].astype(str).str[:4]

CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
